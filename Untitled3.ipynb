```python
import pandas as pd
import numpy as np
from scipy.stats import boxcox
from statsmodels.tsa.stattools import adfuller
import statsmodels.api as sm
import statsmodels.tsa.api as smt

from arch import arch_model

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path = "C:/Users/Александр/1._Лекция_tsml/Series"
df = pd.read_csv(f"{path}/monthly-sales-of-company-x-jan-6.csv", index_col ='Month', parse_dates = True)
df.head()
```

```python
series = df["Count"]

def test_stationarity(timeseries):
    """функция проверки на стационарность"""
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic', 'p-value', '#Lags Used', 'Number of Observations Used'])
    for [key, value] in dftest[4].items():
        dfoutput['Critical Value (%s)' % key] = value
    return(dfoutput)

def tsplot(y, lags=None, figsize=(14, 8), style='bmh'):
    test_stationarity(y)
    if not isinstance(y, pd.Series):
        y = pd.Series(y)
    with plt.style.context(style):
        plt.figure(figsize=figsize)
        layout = (5, 1)
        ts_ax = plt.subplot2grid(layout, (0, 0), rowspan=2)
        acf_ax = plt.subplot2grid(layout, (2, 0))
        pacf_ax = plt.subplot2grid(layout, (3, 0))
        qq_ax = plt.subplot2grid(layout, (4, 0))

        y.plot(ax=ts_ax, color='blue', label='Or')
        ts_ax.set_title('Original')

        smt.graphics.plot_acf(y, lags=lags, ax=acf_ax, alpha=0.05)
        smt.graphics.plot_pacf(y, lags=lags, ax=pacf_ax, alpha=0.05)
        sm.qqplot(y, line='s', ax=qq_ax)
        
        plt.tight_layout()
    return


p = test_stationarity(series)['p-value']

if p < 0.05:
    print(f'p-value={p}, Ряд стационарен')
else:
    print(f'p-value={p}, Ряд не стационарен')

tsplot(series)
```

```python
#Подберем параметры для GARCH

# Список для хранения значений AIC и BIC
results = []

# Перебор различных параметров p и q
for p in range(1, 5):
    for q in range(1, 5):
        model = arch_model(df['Count'], vol='Garch', p=p, q=q)
        result = model.fit(disp="off")
        results.append((p, q, result.aic, result.bic))

# Преобразуем в DataFrame для удобного анализа
results_df = pd.DataFrame(results, columns=['p', 'q', 'AIC', 'BIC'])
print(results_df)

# Находим наилучшие параметры с минимальными AIC и BIC
best_aic = results_df.loc[results_df['AIC'].idxmin()]
best_bic = results_df.loc[results_df['BIC'].idxmin()]

print(f'Best AIC: p={best_aic["p"]}, q={best_aic["q"]}, AIC={best_aic["AIC"]}')
print(f'Best BIC: p={best_bic["p"]}, q={best_bic["q"]}, BIC={best_bic["BIC"]}')
```

```python
# Обучаем модель с лучшими параметрами
optimal_model = arch_model(df['Count'], dist = 'StudentsT', vol='Garch', p=int(best_aic['p']), q=int(best_aic['q']))
optimal_result = optimal_model.fit(disp="off")

# Вывод результатов
print(optimal_result.summary())
```

```python
# Предсказание
horizon=100
pred = optimal_result.forecast(horizon=horizon, method='simulation', simulations=1000)
```

```python
#Строим предсказанный временной ряд
with plt.style.context('bmh'):
    plt.figure(figsize=(10,4))
    ax = plt.axes()
    plt.plot(pred.variance.values[-1, :], color='blue', label='Forecast')
    plt.legend()
    plt.show()
```
